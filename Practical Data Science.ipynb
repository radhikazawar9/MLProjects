```python
# 7) job vs target_subscribe

jobs = bank['job'].value_counts()
m_height = jobs.values.tolist()
jobs.axes 
jobs_labels = jobs.axes[0].tolist() 
yes_cc = [] #yes Subscription list
no_cc = []    #no Subscription list
for x in jobs_labels:
    size = len(bank[bank['job'] == x].index)
    subs = len(bank[(bank['job'] == x) & (bank['target_subscribe'] == 'yes')].index)
    yes_cc.append(subs)
    no_cc.append(size-subs)
                        
ind = np.arange(12) 
width = 0.40
fig, ax = plt.subplots(figsize=(12,7))
yes_bars = ax.bar(ind, yes_cc , width, color='#ADFF2F')
no_bars = ax.bar(ind+width, no_cc , width, color='#DA70D6')


ax.set_xlabel("Types Of Jobs",fontsize=20)
ax.set_ylabel('Frequency',fontsize=20)
ax.set_title('PLOT 1-Subscription Based on Job Type',fontsize=22)
ax.set_xticks(ind + width / 2) 
ax.set_xticklabels(('admin', 'blue-collar','technician','services','  management','retired','  self-employed','  entrepreneur','unemployed','housemaid','student','unknown'),
                  fontsize = 12)
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
ax.legend((yes_bars,no_bars),('yes','no'),fontsize=17)

def autolabel(rects,fontsize=14):
    
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1*height,'%d' % int(height),
                ha='center', va='bottom',fontsize=fontsize)        
autolabel(yes_bars, 10)
autolabel(no_bars, 10)
plt.show()
print(yes_cc)
print(no_cc)
```

```python
#Task 3: Data Modelling

##1) CLASSIFICATION :  
from sklearn.model_selection import train_test_split

## 50% for training and 50% for testing
X_train1, X_test1, Y_train1, Y_test1 = train_test_split(Data,target,test_size=0.50,random_state=999)
print X_train1.shape
print Y_train1.shape
print X_test1.shape
print Y_test1.shape

## 60% for training and 40% for testing
X_train2, X_test2, Y_train2, Y_test2 = train_test_split(Data,target,test_size=0.40,random_state=99)
print X_train2.shape
print Y_train2.shape
print X_test2.shape
print Y_test2.shape

## 80% for training and 20% for testing
X_train3, X_test3, Y_train3, Y_test3 = train_test_split(Data,target,test_size=0.20,random_state=999)
print X_train3.shape
print Y_train3.shape
print X_test3.shape
print Y_test3.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
import sklearn.metrics as metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
from sklearn import feature_selection as fs
from sklearn.neighbors import KNeighborsClassifier
```

```python
#finding value of k for k-nearest neighbor(on 50 : 50 split of data)
scores = []
for each in range(1,10):
    knn_n = KNeighborsClassifier(n_neighbors = each)
    knn_n.fit(X_train1,Y_train1)
    scores.append(knn_n.score(X_train1,Y_train1))
    
plt.plot(range(1,10),scores)
plt.title("k-value for 50-50 split")
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.show()
```

```python
#Feature Selection for K-nearest neighbour algorithm
num_features = 5
model_rfi = RandomForestClassifier(n_estimators=100)
model_rfi.fit(Data, target)
fs_indices_rfi = np.argsort(model_rfi.feature_importances_)[::-1][0:num_features]

best_features_rfi = bank.columns[fs_indices_rfi].values
best_features_rfi

feature_importances_rfi = model_rfi.feature_importances_[fs_indices_rfi]
feature_importances_rfi

y_pos = np.arange(len(best_features_rfi))
plt.bar(y_pos, feature_importances_rfi, align='center', alpha=0.5)
plt.xticks(y_pos, best_features_rfi, rotation = 90)
plt.ylabel('Importance')
plt.title('Random forest: Selected important feature')
plt.show()
```

```python
#K-Nearest Neighbour alogorithm on 50 : 50 split of data
knn_classifier = KNeighborsClassifier(n_neighbors=3, p=2)
fit1 = knn_classifier.fit(X_train1, Y_train1);
knn_50_50 = knn_classifier.score(X=X_test1, y=Y_test1)

predicted1 = fit1.predict(X_test1)
cm = confusion_matrix(Y_test1, predicted1)

f, ax = plt.subplots(figsize =(4,4))
sns.heatmap(cm,annot = True,linewidths=0.5,linecolor="black",fmt = ".0f",ax=ax)
plt.title("PLOT 1-Confusion Matrix 50-50 split")
plt.xlabel("Predicted y values")
plt.ylabel("Real y values")
plt.show()

print("Precision: ", precision_score(Y_test1, predicted1))
print("Recall: ", recall_score(Y_test1, predicted1))
print("F1-score: ",f1_score(Y_test1, predicted1))
print("Classification Accuracy: ",metrics.accuracy_score(Y_test1, predicted1))
print("Classification Error: ", 1 - metrics.accuracy_score(Y_test1, predicted1))
```
